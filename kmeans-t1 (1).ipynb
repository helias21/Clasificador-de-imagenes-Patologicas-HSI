# Modelo de Distribucion por K-Means

Se importan librerias

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
from matplotlib.image import imread

```

Se importa imagen Negativa por cancer

```python
img = imread('032370-20x-roi3.jpg')
```

Se revisan parametros de la imagen (paso no obligatorio, solo para efectos de aprendizaje)

```python
plt.figure(figsize=(12,6))
plt.imshow(img)
```

```python
img=img/255.0
```

```python
img.shape
```

Se aplica el modelo

```python
X=img.reshape(-1,3)
```

```python
X.shape
```

```python
km=KMeans(n_clusters=2)
km.fit(X)
```

```python
img_seg=km.cluster_centers_
```

```python
img_seg
```

```python
img_seg=img_seg[km.labels_]
```

```python
img_seg=img_seg.reshape(img.shape)
```

```python
plt.figure(figsize=(12,6))
plt.imshow(img_seg)
```

Se repite el proceso, esta vez tomando una imagen positiva por cancer

```python
img=imread('040037-20x-roi4.jpg')
```

```python
plt.figure(figsize=(12,6))
plt.imshow(img)
```

```python
img=img/255.0
```

```python
X=img.reshape(-1,3)
```

```python
km=KMeans(n_clusters=2)
km.fit(X)
```

```python
img_seg=km.cluster_centers_
```

```python
img_seg=img_seg[km.labels_]
```

```python
img_seg=img_seg.reshape(img.shape)
```

```python
plt.figure(figsize=(12,6))
plt.imshow(img_seg)
```
